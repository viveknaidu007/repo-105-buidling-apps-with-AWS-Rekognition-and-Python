```python
import boto3

def analyze_video(bucket_name, video_name):
  rekognition_client = boto3.client('rekognition')

  response = rekognition_client.start_label_detection(
      Video={
          'S3Object': {
              'Bucket': bucket_name,
              'Name': video_name
          }
      },
      NotificationChannel={
          'RoleArn': 'arn:aws:iam::381492233747:role/vivek_role',
          'SNSTopicArn': 'arn:aws:sns:ap-south-1:381492233747:shot_detect'
      }
  )

  job_id = response['JobId']
  print('Started video analysis job with job id: {}'.format(job_id))

  # Asynchronous task to get results when the job completes
  get_label_detection_results(rekognition_client, job_id)

def get_label_detection_results(rekognition_client, job_id):
  waiter = rekognition_client.get_waiter('job_status_completed')
  waiter.wait(JobId=job_id)

  response = rekognition_client.get_label_detection(JobId=job_id)

  print('Label detection results for video:')
  for label in response['Labels']:
      print('  Timestamp: {}'.format(label['Timestamp']))
      print('  Label Name: {}'.format(label['Label']['Name']))
      print('  Confidence: {:.2f}%'.format(label['Confidence']))

# Replace with your bucket name, video name, role ARN, and SNS topic ARN
analyze_video('vivekp', 'nature.mp4')
```
